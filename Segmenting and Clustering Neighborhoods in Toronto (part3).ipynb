- Import pacakges
- Read html and get the first table
- Cleanup data and rename columns

```python
import lxml
import pandas as pd
import html5lib
data= pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0] 
df=data.loc[data['Borough']!='Not assigned',:]
df.columns=['PostalCode','Borough','Neighborhood']
df.head()
```

- Test if any Neighborhood has value as 'Not assigned'
- Get the shape of dataframe


```python
df[df['Neighborhood']=='Not assigned']
df.shape
```

- Import data

```python
lat_lon_data = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon_data.columns=['PostalCode','Latitude','Longitude']
lat_lon_data.head()
```

```python
df2 = pd.merge(df,lat_lon_data,on='PostalCode')
df2
```

- Get Toronto data

```python
df_toronto=df2[df2['Borough'].str.contains('Toronto',regex=False)]
df_toronto
```

- Visualizing all the Neighbourhoods of the above data frame using Folium

```python
import folium # plotting library
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighborhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

- Use Kmeans

```python
from sklearn.cluster import KMeans
k=6
toronto_clustering = df_toronto.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_toronto.insert(0, 'Labels', kmeans.labels_)
df_toronto.head()
```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighborhood'], df_toronto['Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
